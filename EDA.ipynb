```python
import json

def load_manifest(file_path):
    with open(file_path, 'r') as file:
        data = [json.loads(line.strip()) for line in file if line.strip()]
    return pd.DataFrame(data)

df_sv_manifest = load_manifest("D:/Fall4/Independent Study/svarah/svarah_manifest.json")
df_sv_l1_manifest = load_manifest("D:/Fall4/Independent Study/svarah/saa_l1_manifest.json")
```

```python
from scipy.stats import chi2_contingency

def hmap(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(k, r) - 1)))

dist_job_qual = pd.crosstab(df_svarah_stats['highest_qualification'], df_svarah_stats['job_category'])
v_value = hmap(dist_job_qual)

plt.figure(figsize=(12, 6))
sns.heatmap(dist_job_qual, annot=True, fmt="d", linewidths=.5, cmap="YlGn", edgecolor='black')
plt.title(f'HEATMAP' , fontsize=20, color="#4daf4a")
plt.xlabel('Job Category', fontsize=14, color="#ff0000")
plt.ylabel('Highest Qualification', fontsize=14, color="#ff0000")
plt.tight_layout()
plt.show()
```
